```python
#Объедините датасеты (кроме students.scv) в один общий для дальнейшей работы. Отдельный датасет создайте для расчета 
#среднего возраста студентов курсов (отдельный датасет со students.scv нужен, чтобы в основном не потерять значения 
#в основном датасете из-за того, что не по всем студентам есть анкетные данные о дате рождения).
'''импортируем модули, выгружаем каждый файл в отдельный датасет'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax
from pandas.plotting import register_matplotlib_converters


ds1 = pd.read_csv("./Data_1/Data/course_contents.csv", encoding = "UTF-8")
ds2 = pd.read_csv("./Data_1/Data/courses.csv", encoding = "UTF-8")
ds3 = pd.read_csv("./Data_1/Data/progresses.csv", encoding = "UTF-8")
ds4 = pd.read_csv("./Data_1/Data/progress_phases.csv", encoding = "UTF-8")
stud = pd.read_csv("./Data_1/Data/students.csv", encoding = "UTF-8")
stud
```

```python
'''Через два промежуточных датасета, путём поочерёдного слияния исходных датасетов, создаём датасет dat. 
Удаляем ненужные столбцы.'''
temp1 = ds4.merge(ds3, how='left', left_on='progress_id', right_on='id')
temp2 = temp1.merge(ds2, how='left', left_on='course_id', right_on='id')
dat = temp2.merge(ds1, how='left')
dat = dat.drop(['id_x', 'id_y', 'Unnamed: 0'], axis=1)
dat
```

```python
#общее количество курсов в датасете найдем с помощью метода unique()
len(dat.course_id.unique())
```

```python
#количество модулей на каждом курсе найдем с помощью группировки датасета по столбцу title и метода nunique()
dat.groupby('title').module_title.nunique()
```

```python
#количество уроков в каждом модуле на каждом курсе найдем с помощью группировки по столбцам title и module_title и метода nunique()
dat.groupby(['title', 'module_title']).lesson_title.nunique().head(40)
```

```python
#медианное количество уроков в модуле на каждом курсе получим группировокй предыдущего результата по столбцу title и метода median()
dat.groupby(['title', 'module_title']).lesson_title.nunique().groupby(['title']).median()
```

```python
#количество учеников на каждом курсе найдем аналогично п.2
dat.groupby(['title']).student_id.nunique()
```

```python
#минимальный, максимальный, средний, медианный возраст студентов
'''Удалим строки с пропусками даты рождения, 
вычислим возраст студентов в годах, путём вычитания дня рождения студента от даты 22 июля 2019(дата создания файлов к курсовой)
в формате datetime и деления на 365(получается несущественная погрешность из-за високосных годов), 
и занесём возраст в столбец 'age'(хранит количество полных лет). Затем удалим всех студентов младше 14 и старше 90 лет 
(допускаем, что студентов младше 14 лет и старше 90 лет не существует, а указанные даты рождения - ошибки).
После этого можем спокойно расчитать требуемые параметры.'''

from datetime import datetime

def age_calc(date_str):
    countdown_date = '2019-07-22'
    countdown_date = datetime.strptime(countdown_date, '%Y-%m-%d').date()
    date_object = datetime.strptime(date_str, '%Y-%m-%d').date()
    return ((countdown_date - date_object).days // 365) 

stud.dropna(subset=['birthday'], inplace=True)
stud["age"] = stud.loc[:, 'birthday'].apply(age_calc)
stud = stud.drop(stud.loc[(stud['age'] < 14) | (stud['age'] > 90)].index)

stud_median = stud.describe().loc["50%"]["age"]
stud_min = stud.describe().loc["min"]["age"]
stud_max = stud.describe().loc["max"]["age"]
stud_mean = stud.describe().loc["mean"]["age"]

print("min={}, max={}, mean={}, median={}".format(stud_min, stud_max, np.around(stud_mean, decimals=2), stud_median))
```

```python
#минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
'''Соеденим таблицу студентов с таблицей соединяющей студентов и курсы, приведём получившуюся таблицу в порядок.
Затем, посчитаем необходимые характеристики и получим их в виде списка. После этого, для красивого вывода, разобьем список на
4 колонки и склеим получившийся фрейм из четырёх колонок с изначальным фреймом, удалив оттуда колонку со списком значений.
Наконец, соеденим итоговы фрейм с таблицей курсов, чтобы видеть название курса и сферу.'''

def my_describe(df):
    df_median = df.describe().loc["50%"]["age"]
    df_min = df.describe().loc["min"]["age"]
    df_max = df.describe().loc["max"]["age"]
    df_mean = df.describe().loc["mean"]["age"]
    return df_min, df_max, np.around(df_mean, decimals=2), df_median


stud_on_curs = stud.merge(ds3, how='left', left_on='id', right_on='student_id').drop(['id_y', 'student_id'], axis=1).rename(columns={'id_x' : 'id'})
temp = stud_on_curs.groupby(['course_id']).apply(my_describe) #.to_frame('describe_list')
temp = pd.DataFrame({'course_id':temp.index, 'describe_list':temp.values})
temp2 = pd.DataFrame(temp.describe_list.tolist(), columns=['min', 'max', 'mean', 'median'])
temp3 = temp.join(temp2).drop('describe_list', axis=1)
ds2.merge(temp3, how='left', left_on='id', right_on='course_id').drop(['Unnamed: 0', 'course_id'], axis=1)
```

```python
#Постройте bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы.

stud_number = dat.groupby(['title']).student_id.nunique()
fig, subplot = plt.subplots()
subplot.bar(stud_number.index, stud_number)
plt.xlabel('Courses')
plt.ylabel('Students')
plt.title('Number of students')
plt.xticks(stud_number.index, rotation='vertical')
plt.show()
```

```python
#Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов на каждом
#курсе. График должен иметь заголовок. Значения должны быть отсортированы. Цвет столбцов должен содержать 
#информацию о сфере, к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. 
#На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно
'''Зададим функцию "field_color", которая возвращает определенный цвет в зависимости от значения параметра 'field'
в передаваемой строке. Применим её при указании аргумента 'color' при создании линий в методе barh(). 
Затем, после стандартных действий при отрисовке графика, создадим легенду. Легенда будет некорректна, поэтому нам надо 
отдельно переопределить линии(handles) и надписи(lables) для неё. Уберём повторы линий и вызовем легенду ещё раз.
'''
def field_color(row):
    colors = ['orange', 'teal', 'blue', 'red']
    if row['field'] == 'Business':
        return colors[0]
    elif row['field'] == 'Design':
        return colors[1]
    elif row['field'] == 'Development':
        return colors[2]
    elif row['field'] == 'Marketing':
        return colors[3]

    
stud_number = dat.groupby(['title', 'field']).student_id.nunique().to_frame().sort_values('student_id').reset_index()
fig, subplot = plt.subplots()
for i in stud_number.index:
    subplot.barh(stud_number.title[i], stud_number.student_id[i], alpha=0.5, color=field_color(stud_number.iloc[i]), \
                 label=stud_number.iloc[i].field)
plt.xlabel('Students')
plt.ylabel('Courses')
plt.title('Number of students')
subplot.spines['right'].set_visible(False)
subplot.spines['top'].set_visible(False)
plt.axvline(stud_number.student_id.median(), color='green', linewidth=2, label='median')

plt.legend()
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). 
#Считать дату начала прохождения курса студентом по дате начала первой домашней работы.
'''Выделим из общего датафрейма только те строки, которые соотсветсвуют началу выполнения первого задания первого модуля 
каждого курса. Удалим все ненужные столбцы. Выделим из столбца "start_date" первые 10 символов, соответсвующих году 
месяцу и дню. Затем преобразуем значения этого столбца из формата строки в формат Datetime.
'''
def start_date_format(str):
    return str[0:10]

growth_stud = dat[(dat['module_number'] == 1) & (dat['lesson_number'] == 1)] \
    .drop(['module_number', 'lesson_number', 'finish_date', 'student_id', 'course_id', 'status', 'field', 'module_title', 'lesson_title', 'lesson_token', 'is_video', 'is_homework'], axis=1)
growth_stud['start_date'] = growth_stud.loc[:, 'start_date'].apply(start_date_format)
growth_stud['start_date'] = pd.to_datetime(growth_stud['start_date'], format='%Y-%m-%d')
growth_stud.set_index('title', inplace=True)
growth_stud
```

```python
'''Затем, создадим датафрейм "zero_dates", содержащий в строках интервалы длинной в месяц, начиная с марта 2016 года 
и заканчивая июлем 2019 года. Создадим список "list_of_titles" с названиями всех курсов. Опишем функцию "expand_grid", 
которая формирует датафрейм, в котором содержатся сочетания всех курсов по всем месяцам. 
Принимает в качестве аргумента словарь из заголовков столбцов и значений, по которым нужно сделать сочетания.
Применим эту функцию с словарю из списка "list_of_titles" и датафрейма "zero_dates", получив датасет month_set.
'''
import itertools

def expand_grid(data_dict):
    rows = itertools.product(*data_dict.values())
    return pd.DataFrame.from_records(rows, columns=data_dict.keys())

zero_dates = pd.date_range(start='3/1/2016', end='07/31/2019', freq='M')
list_of_titles = ds2['title'].tolist()
month_set = expand_grid({'title': list_of_titles, 'start_date': zero_dates,})
month_set
```

```python
'''Опишем функцию "growth_of_stud", которая формирует датасет, показывающий, сколько новых студентов пришло на курс в каждом
месяце. Принимает на вход датафрейм и название курса. Выделяет данные только по курсу, соотвтевующему названию,
группирует их на периоды помесячно и подсчитывает количество студентов. Затем, формирует аналогичный датасет из датасета 
"month_set", но с нулевыми занчениями. После этого, объединяет два датасета с помощью метода "concat" таким образом, 
чтобы значения прироста студентов заменили нулевые значения из преобразованного "month_set", а остальные нулевые значения
остались на месте.
Применим функцию "growth_of_stud" к датафрейму "growth_stud" и к каждому элементу списка "list_of_titles" и выведем на экран, получив расчитанный прирост 
студентов помесячно, для каждого курса, в заданный пероид.
'''

def growth_of_stud(input_df, title):
    temp_df_growth_stud = input_df.loc[title]
    temp_df_growth_stud = temp_df_growth_stud.groupby(temp_df_growth_stud.start_date.dt.to_period("M")) \
        .agg('count').drop(['start_date'], axis=1) \
        .reset_index() \
        .rename(columns={'progress_id' : 'count'})

    temp_df_month_set = month_set.set_index('title').loc[title]
    temp_df_month_set = temp_df_month_set.groupby(temp_df_month_set.start_date.dt.to_period("M")) \
        .agg('count') \
        .drop(['start_date'], axis=1) \
        .reset_index()
    temp_df_month_set['count'] = 0

    result = pd.concat([temp_df_growth_stud, temp_df_month_set], join='outer',ignore_index=True, verify_integrity=True) \
        .sort_values(by=['start_date']) \
        .groupby('start_date') \
        .agg('sum') \
        .reset_index() \
        .rename(columns={'start_date' : 'year_month'})
    result['year_month'] = result.year_month.values.astype('datetime64[M]')
    return result


for i in list_of_titles:
    print(i, '\n', growth_of_stud(growth_stud, i))
```

```python
#Рассчитать количество прогрессов по выполнению домашних работ в каждом месяце за всю историю 
#(каждый месяц в диапазоне от марта 2016 до июля 2019 включительно) для каждого курса. 
#Учитывать, что выполнение домашнего задания может перетекать из одного месяца в другой 
#(такие дз надо включать в общее число прогрессов для всех месяцев, которые покрывает срок выполнения этих дз)
'''
Удалим ненужные столбцы из исходного датафрейма и присовим результат датафрейму 'growth_progress'. 
Преобразуем столбцы 'start_date' и 'finish_date' в формат DdateTime. Ячейки в столбце 'finish_date',
заполненные значениями Nan, соответсвтующими невыполненому заданию, заполним значением 2019-07-22. Переведём оба столбца в 
формат Period помесячно. Создадим столбец 'delta_date' со значениями разности значений столбца 'finish_date' и 'start_date'.
РАсчёты будем производить в формате int, чтобы получить целочисленные значения в днях. 
Зададим список 'list_of_multi_progress'.
В цикле пройлёмся по всем строкам датафрейма 'growth_progress'. Если значение столбца 'delta_date' на текущей итерации 
больше нуля, т.е. домашнее задание выполнялось в период, занимающий более одного месяца, то попадём ещё в один цикл
по количеству месяцев, которое потребовалось на выполнение текущего ДЗ. В каждой итерации заносим в список 
'list_of_multi_progress' словарь, состоящий из ключа 'start_date' и значение первого месяца выполнения плюс 
значение счётчика. Таким образом мы разворачиваем каждое многомесячное ДЗ на отдельные месяцы.
Соеденим значения датафрейма 'growth_progress' и списка 'list_of_multi_progress' через метод concat().
Заполним значения Nan в столбце 'delta_date' нулями и удалим все ненулевые значения, удалив таким образом многомесячные
прогрессы, которые мы уже развернули и добавили в конец датафрейма. Удалим, также, ненужные столбцы.
'''
def start_date_format(str):
    return str[0:10]

growth_progress = dat.drop(['module_number', 'lesson_number', 'student_id', 'course_id', 'status', 'field', 'module_title', 'lesson_title', 'lesson_token', 'is_video', 'is_homework'], axis=1)
growth_progress['start_date'] = growth_progress.loc[:, 'start_date'].apply(start_date_format)
growth_progress['start_date'] = pd.to_datetime(growth_progress['start_date'], format='%Y-%m-%d')
growth_progress['finish_date'] = growth_progress['finish_date'].fillna(value='2019-07-22')
growth_progress['finish_date'] = growth_progress.loc[:, 'finish_date'].apply(start_date_format)
growth_progress['finish_date'] = pd.to_datetime(growth_progress['finish_date'], format='%Y-%m-%d')
growth_progress = growth_progress.set_index('start_date').to_period('M').reset_index()
growth_progress = growth_progress.set_index('finish_date').to_period('M').reset_index()

growth_progress['delta_date'] = growth_progress['finish_date'].astype('int') - growth_progress['start_date'].astype('int')
list_of_multi_progress=[]

for i in growth_progress.index:
    if growth_progress['delta_date'][i] > 0:
        for j in range(growth_progress['delta_date'][i] + 1):
            list_of_multi_progress.append({'start_date': growth_progress['start_date'][i] + j, 'title': growth_progress['title'][i]}) 

growth_progress = pd.concat([growth_progress, pd.DataFrame(list_of_multi_progress)], ignore_index=True, sort=False)
growth_progress['delta_date'] = growth_progress['delta_date'].fillna(0)
growth_progress.drop(growth_progress[growth_progress['delta_date'] > 0].index, inplace=True)
growth_progress = growth_progress.drop(['finish_date', 'progress_id'], axis=1)
growth_progress
```
