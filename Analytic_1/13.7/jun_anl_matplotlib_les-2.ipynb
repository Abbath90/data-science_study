## Простой график Matplotlib

Добавим на график показатель загрязнённости оксидом азота, для этого:
* достанем соответствующие столбцы из DataFrame: сам показатель и временные отметки, когда был зафиксирован этот показатель
* передадим их в метод `.plot()`

Подготовка данных

```python
import pandas as pd
from datetime import datetime

air_df = pd.read_csv('./data/air_quality.tar.gz', compression='gzip', sep=';', encoding='utf8')
air_df = air_df[air_df.iloc[:,0].notnull()]

# формируем столбец с таймстемпом
air_df['ts'] = (air_df.iloc[:,0]+ ' ' + air_df.iloc[:,1]).map(
    lambda x: datetime.strptime(x.replace('.',':'), "%d/%m/%Y %H:%M:%S")
)

air_df.head()
```

Отрисовка графика

```python
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import register_matplotlib_converters
%matplotlib inline

register_matplotlib_converters()
TOP=25
# доступ к Figure и Subplot
fig, subplot = plt.subplots()
# формируем значения по оси  X
x_ticks = air_df['ts'].values[:TOP]
# формируем значения по оси Y
y_ticks = air_df['PT08.S1(CO)'].values[:TOP]
# график функции
subplot.plot(x_ticks, y_ticks)
plt.show()
```

График выглядит отвратительно - оси не подписаны, в подписях на координатных осях какая-то каша.

Тем не менее, основная цель достигнута - данные визуализированы! Более детальный анализ будет проведён в следущих уроках.
